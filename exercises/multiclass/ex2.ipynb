# Multiclass - Exercise 2

Create a dataset with three classes, each class has N = 500 objects, represented by two variables X1 and X2 according to the following requirements (seed = 37):

The feature $X^1$ is distributed as follows:
- $X^1_1$ is an exponential distribution with a scale parameter of 3
- $X^1_2$ is a normal distribution with a mean of 5 and a standard deviation of 3
- $X^1_3$ is a normal distribution with a mean of -1 and a standard deviation of 4

The feature $X^2$ is distributed as follows:
- $X^2_1$ is a normal distribution with a mean of 2 and a standard deviation of 1.5
- $X^2_2$ is a normal distribution with a mean of 2 and a standard deviation of 2
- $X^2_3$ is a normal distribution with a mean of 4 and a standard deviation of 1

Each class has its label $Y$ that can be $1$, $2$ or $3$

```python
import numpy as np 

# set seed
np.random.seed(37)

# number of samples
N = 500

# generate 1st class
X1_1 = - np.random.exponential(scale=3, size=N)
X1_2 = np.random.normal(loc=-2, scale=1.5, size=N)

# generate 2nd class
X2_1 = np.random.normal(loc=5, scale=3, size=N)
X2_2 = np.random.normal(loc=-2, scale=3, size=N)

# generate 3rd class
X3_1 = np.random.normal(loc=-1, scale=4, size=N)
X3_2 = np.random.normal(loc=4, scale=1, size=N)

# labels
Y1 = np.ones(N)
Y2 = 2 * np.ones(N)
Y3 = 3 * np.ones(N)
```

Plot the generated dataset using the <tt>scatter</tt> function

```python
import matplotlib.pyplot as plt

# first class
plt.scatter(X1_1, X1_2)

# second class
plt.scatter(X2_1, X2_2)

# third class
plt.scatter(X3_1, X3_2)
```

Create a matrix of points $X$ (each row is an object) and a vector $y$ of labels. Remember that we need the coordinate $X_0$ for the bias term (all ones).

```python
# bias vector
X0 = np.ones(3 * N)

# feature X1
X1 = np.concatenate((X1_1, X2_1, X3_1))

# feature X2
X2 = np.concatenate((X1_2, X2_2, X3_2))

# dataset X
X = np.stack((X0, X1, X2), axis = 1)

# vector of labels
y = np.concatenate((Y1, Y2, Y3))
```

Create a vector of labels <tt>y_class</tt> that contains 1 for each object belonging to the class under study (positive class) and -1 for all the other objects. The positive class is the third.

```python
# choose label as positive class
positive_class = 3

# new vector of labels
y_class = - np.ones_like(y)

# masking
y_class[ y == positive_class ] = 1
```

Using <tt>plt.scatter</tt> plot the new binary classification problem

```python
plt.scatter(X[:, 1], X[:, 2], c=y_class)
```

## Error function

Write the function <tt>sse(X, y, w)</tt> that takes the data matrix $X$, the labels $Y$, and the vector of parameters $w$ and computes sum of squared errors defined as $\sum\left( Y - X\cdot w \right)^2$

```python
def sse(X : np.matrix, y : np.array, w : np.array) -> float:
    return np.sum( np.square( y - np.dot( X, w ) ) )
```

## Learning

After defining the vector of weights as a normal distribution with $\mu = 1$ and $\sigma = 3$, start training the model by using $\eta = 20^{-4}$ and the maximum epochs to 200. 

For each epoch save the sum of squared errors in the <tt>sse_epoch</tt> vector and the number of objects misclassified in the vector <tt>errors_epoch</tt>.


```python
# set up the weight vector
w = np.random.normal(loc=1, scale=3, size=X.shape[1])

# learning rate
eta = 1e-4

# epochs
max_epochs = 300
current_epoch = 0

# errors
sse_epoch = np.zeros(max_epochs)
errors_epoch = np.zeros(max_epochs)

while current_epoch < max_epochs:

    # sum of squared errors
    sse_epoch[current_epoch] = sse(X, y_class, w)

    for i in range(y_class.size):
        output = np.sign(np.dot(X[i, :], w))

        if output * y_class[i] < 0 :
            w += eta * X[i, :] * y_class[i]
            errors_epoch[current_epoch] += 1

    current_epoch += 1
```

Display raw the plots containing the solution found by the PLA

```python
# datapoints
plt.scatter(X[:, 1], X[:, 2], c=y_class)

# plot the solution
plt.plot([ np.min((X1_1, X2_1, X3_1)), np.max((X1_1, X2_1, X3_1)) ],
         [ np.min((X1_1, X2_1, X3_1)) * (- w[1] / w[2]) - (w[0] / w[2]), np.max((X1_1, X2_1, X3_1)) * (- w[1] / w[2]) - (w[0] / w[2]) ],
         color = "green")
```

Now display the two error plots using the two error vectors

```python
# create subplot
fig, axs = plt.subplots(1, 2, figsize=(15, 5))

# sum of squared errors
axs[0].plot(range(sse_epoch.size), sse_epoch)

# errors per epoch
axs[1].plot(range(errors_epoch.size), errors_epoch)
```
